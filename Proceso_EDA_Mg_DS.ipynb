### 3.5 Separar en facetas

```r
# Según una sola variable
ggplot(data = mpg) +
 geom_point(mapping = aes(x = displ, y = hwy)) +
 facet_wrap(~ class, nrow = 2)
```

```r
# Según combinación de dos variables.
ggplot(data = mpg) +
 geom_point(mapping = aes(x = displ, y = hwy)) +
 facet_grid(drv ~ cyl)
```

Los graficos obtenidos no tienen mucho sentido. R tratara de seperar las facetas por cada valor unico de la variable continua. A su vez si tengo muchos valores unicos, R generara un error.

Las celdas vacías significan que no hay observaciones en los datos que tengan esa combinación única de valores. La trama es similar a la original, solo que cada faceta solo parece tener un único punto de datos.

3. ¿Qué grafica el siguiente código? ¿Qué hace . ?

```r
ggplot(data = mpg) +
  geom_point(mapping = aes(x = displ, y = hwy)) +
  facet_grid(drv ~ .)
```

```r
ggplot(data = mpg) +
  geom_point(mapping = aes(x = displ, y = hwy)) +
  facet_grid(. ~ cyl)
```

EL simbolo de punto '.' actúa como un marcador de posición para ninguna variable o no indicar ninguna variable. En facet_grid(), esto da como resultado un gráfico facetado en una sola dimensión (1 por N o N por 1) en lugar de una cuadrícula o grilla.

4. Mira de nuevo el primer gráfico en facetas presentado en esta sección:

```r
ggplot(data = mpg) +
  geom_point(mapping = aes(x = displ, y = hwy)) +
  facet_wrap(~class, nrow = 2)
```

¿Cuáles son las ventajas de separar en facetas en lugar de aplicar una estética de color? ¿Cuáles son las desventajas? ¿Cómo cambiaría este balance si tuvieras un conjunto de datos más grande?

```r
ggplot(data = mpg) +
  geom_point(mapping = aes(x = displ, y = hwy, color = class))
```

Las facetas dividen los datos en cuadrículas separadas y visualizan mejor las tendencias dentro de cada faceta individual. La desventaja es que, al hacerlo, es más difícil visualizar la relación general entre las facetas. La estética del color está bien cuando su conjunto de datos es pequeño, pero con conjuntos de datos más grandes, los puntos pueden comenzar a superponerse entre sí. En esta situación con un gráfico en colores, la marcación puede no ser suficiente debido a la estética del color adicional.

5. Lee ?facet_wrap. ¿Qué hace nrow? ¿Qué hace ncol? ¿Qué otras opciones controlan el diseño de los paneles individuales? ¿Por qué facet_grid() no tiene argumentos nrow y ncol?

* **nrow** establece cuántas filas tendrá el gráfico facetado.
* **ncol** establece cuántas columnas tendrá el gráfico facetado.
* **as.table** determina la faceta de inicio para comenzar a llenar el gráfico.
* **dir** determina la dirección de inicio para completar el gráfico (horizontal o vertical).
